```python
import numpy as np
import matplotlib.pyplot as plt
import pickle
```

```python
path = '/kaggle/input/analysis/history.pkl'
with open(path, 'rb') as file:
    analysis = pickle.load(file)
```

```python
analysis.keys()
```

```python
epochs = np.arange(len(analysis['Train Loss']))
plt.figure(figsize=(10, 6))
plt.plot(epochs, analysis['Train Loss'], label='Train Loss')
plt.plot(epochs, analysis['Valid Loss'], label='Validation Loss')
plt.title('Train and Validation Loss Over Epochs')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.grid(True)
plt.legend()
plt.show()
```

```python
epochs = np.arange(len(analysis['Train Loss']))
plt.figure(figsize=(10, 6))
plt.plot(epochs, analysis['Train r2'], label='Train r2')
plt.plot(epochs, analysis['Valid r2'], label='Validation r2')
plt.title('Train and Validation Loss Over Epochs')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.grid(True)
plt.legend()
plt.show()
```
